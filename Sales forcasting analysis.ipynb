# Model Making( Linear Regression )

#### Split the dataset into traing and testing dataset with Random state 100% and train_size is 70% and test_size is 30%


```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['TV', 'Radio', 'Newspaper']] 
y = df['Sales'] 
```

```python
x.head(10)
```

```python
y.head()
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7,test_size=0.3, random_state=100)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Standerize the datset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)

x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled 
```

### Model Fitting

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
print("Predictions on Test Data:", y_pred)
```

```python
y_pred.shape
```

```python
m=model.coef_
c=model.intercept_
print("Value of Coefficient or m:",m)
print("Value of Intercept or c:",c)
```

### Find the model is best suitable or not( Model Checking)

```python

from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
rmse = np.sqrt(mse)
```

```python
sse = np.sum((y_test - y_pred) ** 2)
```

```python
k=np.mean(y_pred)
```

```python
tss=np.sum((y_test-k)**2)
```

```python
rss=np.sum((y_test - y_pred) ** 2)
```

```python
r2=1-(rss/tss)
```

```python
### Print all the matrices
print("SSE OR SUM OF SQURE ERROR",sse)
print("\nMSE OR MEAN OF SQURE ERROR:",mse)
print("\nRMSE OR ROOT MEAN OF SQURE ERROR:",rmse)
print("\nR2 VALUE:",r2)
```

### SOME PLOTS

```python
plt.scatter(y_test, y_pred,color="red")
plt.plot(y_test, y_pred, color="Blue", alpha=0.7)
plt.show()
```

## Conclusion:

The model is performing well.

R = 0.867 means that 86.7% of the variance in the target variable is explained by the model. This is strong, especially in real-world data where anything above 0.8 is generally considered good.

RMSE = 1.62 shows that prediction errors are small compared to the overall scale of the data (which ranges from around 6 to 22).

Residual plot (your image) indicates that most predictions are fairly close to the true values, with only a few noticeable errors.
